### Drop Unnecessary Columns

```python
df.drop(df.loc[:, 'AGE_LESS_5':'AGE_85_PLUS'], inplace = True, axis = 1)
df.drop(df.loc[:, 'PCT_HISPANIC':'VACANTUNITS_2010'], inplace = True, axis = 1)
df.drop(df.loc[:, 'NUM_HOUSEHOLDS':], inplace = True, axis = 1)
df = df.drop(['PCT_LESS_18', 'PCT_65_PLUS'], axis=1)
```

### Format Headers

```python
df.columns = df.columns.str.strip().str.lower().str.replace('_', ' ').str.replace(' 2010', "")
df.index.names = ['nbhd id']
df.head(1)
```

## Visualize and Analyze the Data

### Population Distribution Across the Neighborhoods

```python
df.plot.bar(y="population", figsize=(20,5))
```

```python
rdist = df.loc[:, 'hispanic':'two or more']
rdist.head()
```

#### Total Population of Various Races in Denver

```python
rdist.sum()
```

```python
plot = rdist.sum().plot(kind='pie', ylabel='', labeldistance=None, title="Denver's Racial Distribution")
labels = [f'{i}: {round(rdist[i].sum()/rdist.sum().sum()*100, 2)}%' for i in rdist]
plot.legend(loc='center left',bbox_to_anchor=(1.0, 0.5), labels=labels)
```

#### Racial Distribution of Various Neighborhoods

```python
# the table technically already includes this, but I recalculated it for fun
prdist = rdist.div(rdist.sum(axis=1), axis=0)
prdist
```

#### Racial Concentrations

```python
# Which neighborhoods have higher percentage of Hispanics (relative to % in other neighborhoods)
df.loc[prdist[prdist['hispanic'] > prdist['hispanic'].mean()].index]['nbrhd name']
```

```python
# Which neighborhoods have higher percentage of African Americans (relative to % in other neighborhoods):
df.loc[prdist[prdist['black'] > prdist['black'].mean()].index]['nbrhd name']
```

## Age Distribution

#### Population in Neighborhood Compared with Percentage of Neighborhood Population Aged 20-29

```python
df['percent 20-29'] = round(df['age 20 to 29']/df['population']*100, 2)

ax = df.plot.line(x="nbrhd name", y='percent 20-29', label="Percentage Aged 20-29", color="red")
df.plot.bar(x="nbrhd name", y="population", figsize=(20,5), color="green", secondary_y = True, ax = ax, label="Population", title="Neighborhood Population vs. Percentage Aged 20-29")
```

```python
df = df.drop('percent 20-29', axis=1)
```
